```python
import subprocess
import os
import logging
from pathlib import Path
```

```python
logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] %(message)s")
logger = logging.getLogger("FullPipeline")
```

```python
def _get_py(venv: Path):
    # Force venv to be a Path object even if a string was passed
    venv = Path(venv) 
    return str(venv / "Scripts" / "python.exe" if os.name == "nt" else venv / "bin" / "python")
```

```python
def run_step(venv: Path, script: str, args: list):
    venv = Path(venv)
    py = _get_py(venv)
    
    # --- DEBUG CHECKS ---
    if not Path(py).exists():
        raise FileNotFoundError(f"Python NOT FOUND at: {py}")
    if not Path(script).exists():
        raise FileNotFoundError(f"Script NOT FOUND at: {script}")
    # --------------------

    cmd = [py, script] + args
    logger.info(f"Running: {script} in {venv.name}")
    
    result = subprocess.run(cmd, capture_output=True, text=True, encoding='utf-8')
    if result.returncode != 0:
        logger.error(f"Error in {script}:\n{result.stderr}")
        raise RuntimeError(f"Step failed: {script}")
    return result.stdout
```

```python
img_abs = str(Path("../imgs/8.jpg").absolute())
run_step("../venv_render", "sub_render.py", [img_abs])
```
